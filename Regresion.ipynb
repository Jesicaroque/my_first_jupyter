```python
import pandas as pd
import numpy as np
from sklearn import linear_model

df_dep = pd.read_csv('departamentos-pro.csv').astype(str)
df_casa = pd.read_csv('casas-pro.csv').astype(str)

formatito()

dep = pd.DataFrame(df_dep.loc[df_dep['web-scraper-start-url'].str.contains("cumbres"), 'precio'])
dep['metros'] = df_dep.loc[df_dep['web-scraper-start-url'].str.contains("cumbres"), 'metros']

cas = pd.DataFrame(df_casa.loc[df_casa['web-scraper-start-url'].str.contains("cumbres"), 'precio'])
cas['metros'] = df_casa.loc[df_casa['web-scraper-start-url'].str.contains("cumbres"), 'metros']

def formatito():

    #Departamentitos
    df_dep['precio'] = pd.to_numeric(df_dep['precio'].str.replace('MN ', '').str.replace('$', '', regex=False).str.replace(',', '', regex=False), errors = 'coerce')
    df_dep['metros'] = pd.to_numeric(df_dep['mets'].str.replace(' m²', '', regex=True), errors = 'coerce')

    model_dep = linear_model.LinearRegression()

    #Casas
    df_casa['precio'] = pd.to_numeric(df_casa['precio'].str.replace('MN ', '').str.replace('$', '', regex=False).str.replace(',', '', regex=False), errors = 'coerce')
    df_casa['metros'] = pd.to_numeric(df_casa['mets'].str.replace(' m²', '', regex=True), errors = 'coerce')

    model_casa = linear_model.LinearRegression()

    
def departamentos():

    print("Departamentos\n")
    print("Número de filas: ", dep.shape[0])

    dep_70=dep.iloc[0:68]

    precio_depts = pd.DataFrame(dep_70['precio'])
    metros_depts = pd.DataFrame(dep_70['metros'])

    model_dep.fit(metros_depts,precio_depts)
    print("Entrenamiento:\t", model_dep.predict(pd.DataFrame([111])))

    dep_30=dep.iloc[68:]

    precio_depts = pd.DataFrame(dep_30['precio'])
    metros_depts = pd.DataFrame(dep_30['metros'])

    model_dep.fit(metros_depts,precio_depts)
    print("Validación:\t", model_dep.predict(pd.DataFrame([111])))

def casas():

    print("\nCasas\n")
    print("Número de filas: ", cas.shape[0])

    cas_70=cas.iloc[0:1203]

    precio_cas = pd.DataFrame(cas_70['precio'])
    metros_cas = pd.DataFrame(cas_70['metros'])

    model_casa.fit(metros_cas,precio_cas)
    print("Entrenamiento:\t", model_casa.predict(pd.DataFrame([271])))

    cas_30=cas.iloc[1203:]

    precio_cas = pd.DataFrame(cas_30['precio'])
    metros_cas = pd.DataFrame(cas_30['metros'])

    model_casa.fit(metros_cas,precio_cas)
    print("Validación:\t", model_casa.predict(pd.DataFrame([271])))


departamentos()
casas()
```
